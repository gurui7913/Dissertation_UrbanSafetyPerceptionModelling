```python
import pandas as pd
import numpy as np
import os
```

```python
# Read CSV file
# Replace 'your_file.csv' with your actual file path
df = pd.read_csv(r'E:\01_UCL\08_Dissertation\02_DataProcessing\01_SpaceSyntax_London\London_RegressionModel_wgs84.csv')

print("Data Overview:")
print(f"Total rows: {len(df)}")
print("\n" + "="*60)

# Check column names first
print("Column Names:")
print(df.columns.tolist())
print("\n" + "="*60)

# Analyze safer_Trueskill_Scores max and min values
print("Safer TrueSkill Scores Analysis:")
print("-" * 30)

# Maximum value
max_trueskill = df['safer_Trueskill_Scores'].max()
max_trueskill_row = df[df['safer_Trueskill_Scores'] == max_trueskill]

print(f"Maximum Safer TrueSkill Score: {max_trueskill}")
for idx, row in max_trueskill_row.iterrows():
    print(f"Location ID: {row['location_id']}")
    print(f"Longitude: {row['lon']}")
    print(f"Latitude: {row['lat']}")
    print(f"Corresponding INT2K: {row['INT2K']}")
    print(f"Corresponding CH2K: {row['CH2K']}")
    print()

# Minimum value
min_trueskill = df['safer_Trueskill_Scores'].min()
min_trueskill_row = df[df['safer_Trueskill_Scores'] == min_trueskill]

print(f"Minimum Safer TrueSkill Score: {min_trueskill}")
for idx, row in min_trueskill_row.iterrows():
    print(f"Location ID: {row['location_id']}")
    print(f"Longitude: {row['lon']}")
    print(f"Latitude: {row['lat']}")
    print(f"Corresponding INT2K: {row['INT2K']}")
    print(f"Corresponding CH2K: {row['CH2K']}")
    print()

# Collect all extreme values data to avoid output truncation
print("Analyzing extreme values data...")

# INT2K analysis
max_int2k = df['INT2K'].max()
max_int2k_row = df[df['INT2K'] == max_int2k]
min_int2k = df['INT2K'].min()
min_int2k_row = df[df['INT2K'] == min_int2k]

# CH2K analysis  
max_ch2k = df['CH2K'].max()
max_ch2k_row = df[df['CH2K'] == max_ch2k]
min_ch2k = df['CH2K'].min()
min_ch2k_row = df[df['CH2K'] == min_ch2k]

print("="*60)
print("All Extreme Values Detailed Information:")
print("="*60)

# Display key information directly to avoid excessive output causing truncation
print("[INT2K Extreme Values Information]")
print(f"Maximum INT2K: {max_int2k}")
for idx, row in max_int2k_row.iterrows():
    print(f"  Location ID: {row['location_id']}")
    print(f"  Longitude: {row['lon']}, Latitude: {row['lat']}")
    print(f"  Corresponding Safer TrueSkill Score: {row['safer_Trueskill_Scores']}")

print(f"\nMinimum INT2K: {min_int2k}")
for idx, row in min_int2k_row.iterrows():
    print(f"  Location ID: {row['location_id']}")
    print(f"  Longitude: {row['lon']}, Latitude: {row['lat']}")
    print(f"  Corresponding Safer TrueSkill Score: {row['safer_Trueskill_Scores']}")

print("\n" + "-"*40)
print("[CH2K Extreme Values Information]")
print(f"Maximum CH2K: {max_ch2k}")
for idx, row in max_ch2k_row.iterrows():
    print(f"  Location ID: {row['location_id']}")
    print(f"  Longitude: {row['lon']}, Latitude: {row['lat']}")
    print(f"  Corresponding Safer TrueSkill Score: {row['safer_Trueskill_Scores']}")

print(f"\nMinimum CH2K: {min_ch2k}")
for idx, row in min_ch2k_row.iterrows():
    print(f"  Location ID: {row['location_id']}")
    print(f"  Longitude: {row['lon']}, Latitude: {row['lat']}")
    print(f"  Corresponding Safer TrueSkill Score: {row['safer_Trueskill_Scores']}")

print("\n" + "-"*40)
print("[Quick Reference Table]")
print("Safer TrueSkill Score:")
print(f"  Maximum: {max_trueskill} → ID: {max_trueskill_row.iloc[0]['location_id']}")
print(f"  Minimum: {min_trueskill} → ID: {min_trueskill_row.iloc[0]['location_id']}")
print("INT2K:")
print(f"  Maximum: {max_int2k} → ID: {max_int2k_row.iloc[0]['location_id']}")
print(f"  Minimum: {min_int2k} → ID: {min_int2k_row.iloc[0]['location_id']}")
print("CH2K:")
print(f"  Maximum: {max_ch2k} → ID: {max_ch2k_row.iloc[0]['location_id']}")
print(f"  Minimum: {min_ch2k} → ID: {min_ch2k_row.iloc[0]['location_id']}")

print("\n" + "="*60)
print("All data analysis completed!")

# Create DataFrame for extreme values information
print("\nGenerating extreme values CSV file...")
print("-" * 40)

# Collect all extreme values data
extreme_values_data = []

# Safer TrueSkill Score maximum values
for idx, row in max_trueskill_row.iterrows():
    extreme_values_data.append({
        'Metric': 'Safer_TrueSkill_Score',
        'Type': 'Maximum',
        'Value': max_trueskill,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# Safer TrueSkill Score minimum values
for idx, row in min_trueskill_row.iterrows():
    extreme_values_data.append({
        'Metric': 'Safer_TrueSkill_Score',
        'Type': 'Minimum',
        'Value': min_trueskill,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# INT2K maximum values
for idx, row in max_int2k_row.iterrows():
    extreme_values_data.append({
        'Metric': 'INT2K',
        'Type': 'Maximum',
        'Value': max_int2k,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# INT2K minimum values
for idx, row in min_int2k_row.iterrows():
    extreme_values_data.append({
        'Metric': 'INT2K',
        'Type': 'Minimum',
        'Value': min_int2k,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# CH2K maximum values
for idx, row in max_ch2k_row.iterrows():
    extreme_values_data.append({
        'Metric': 'CH2K',
        'Type': 'Maximum',
        'Value': max_ch2k,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# CH2K minimum values
for idx, row in min_ch2k_row.iterrows():
    extreme_values_data.append({
        'Metric': 'CH2K',
        'Type': 'Minimum',
        'Value': min_ch2k,
        'Location_ID': row['location_id'],
        'Longitude': row['lon'],
        'Latitude': row['lat'],
        'Safer_TrueSkill_Score': row['safer_Trueskill_Scores'],
        'INT2K': row['INT2K'],
        'CH2K': row['CH2K']
    })

# Create DataFrame
extreme_values_df = pd.DataFrame(extreme_values_data)

# Display DataFrame
print("Extreme values data preview:")
print(extreme_values_df.to_string(index=False))

# Save as CSV file
# Option 1: Save in current directory
# output_filename = 'extreme_values_analysis.csv'

# Option 2: Specify custom path
output_filename = r'E:\01_UCL\08_Dissertation\04_DataAnalysis\02_极值分析\extreme_values_analysis.csv'

extreme_values_df.to_csv(output_filename, index=False, encoding='utf-8-sig')

print(f"\nCSV file generated: {output_filename}")
print(f"File location: E:\01_UCL\08_Dissertation\04_DataAnalysis\02_极值分析")
print(f"Total extreme value records: {len(extreme_values_df)}")

# Create summary statistics CSV
summary_stats = []
metrics = ['Safer_TrueSkill_Score', 'INT2K', 'CH2K']
values = [(max_trueskill, min_trueskill), (max_int2k, min_int2k), (max_ch2k, min_ch2k)]

for metric, (max_val, min_val) in zip(metrics, values):
    summary_stats.append({
        'Metric': metric,
        'Maximum_Value': max_val,
        'Minimum_Value': min_val,
        'Range': max_val - min_val,
        'Average': (max_val + min_val) / 2
    })

summary_df = pd.DataFrame(summary_stats)
summary_filename = 'summary_statistics.csv'

# Option 2: Specify custom path if needed
# summary_filename = r'C:\Your_Path\summary_statistics.csv'

summary_df.to_csv(summary_filename, index=False, encoding='utf-8-sig')

print(f"Summary statistics file generated: {summary_filename}")

# Display current working directory
print(f"Current working directory: {os.getcwd()}")
print("To specify save path, modify the output_filename variable in the code")

print("\n" + "="*60)
```
