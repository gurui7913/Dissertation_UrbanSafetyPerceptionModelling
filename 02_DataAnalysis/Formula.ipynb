## Mathematical Model Formulations

This research involves four main regression models, described in detail using mathematical formulas:

### 1. **Basic Linear Regression Model**

$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_4 + \varepsilon$$

Where:
- $Y$ = `safer_Trueskill_Scores` (safety perception scores)
- $X_1$ = `INT2K` (spatial integration, standardized)
- $X_2$ = `CH2K` (choice values, standardized)  
- $X_3$ = `green_view_ratio` (green view ratio, standardized)
- $X_4$ = `sky_visibility` (sky visibility, standardized)
- $\beta_0$ = intercept term
- $\beta_1, \beta_2, \beta_3, \beta_4$ = regression coefficients
- $\varepsilon$ = random error term, $\varepsilon \sim N(0, \sigma^2)$

### 2. **Polynomial Regression Model (Degree 2)**

$$Y = \beta_0 + \sum_{i=1}^{4} \beta_i X_i + \sum_{i=1}^{4} \gamma_i X_i^2 + \sum_{i=1}^{3} \sum_{j=i+1}^{4} \delta_{ij} X_i X_j + \varepsilon$$

Expanded form:
$$\begin{aligned}
Y = &\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_4 \\
&+ \gamma_1 X_1^2 + \gamma_2 X_2^2 + \gamma_3 X_3^2 + \gamma_4 X_4^2 \\
&+ \delta_{12} X_1 X_2 + \delta_{13} X_1 X_3 + \delta_{14} X_1 X_4 \\
&+ \delta_{23} X_2 X_3 + \delta_{24} X_2 X_4 + \delta_{34} X_3 X_4 + \varepsilon
\end{aligned}$$

Where:
- $\gamma_i$ = quadratic term coefficients (capturing non-linear relationships, e.g., inverted-U)
- $\delta_{ij}$ = interaction term coefficients (capturing interactions between variables)

### 3. **Enhanced Linear Model with Targeted Interactions**

This strategically designed model examines six theoretically motivated interaction terms: spatial configuration × environmental features (INT2K×green_view_ratio, INT2K×sky_visibility, CH2K×green_view_ratio, CH2K×sky_visibility), visual element interactions (green_view_ratio × sky_visibility), and spatial complexity effects (INT2K×CH2K). This approach enabled examination of conditional relationships where environmental feature effects vary depending on spatial configuration characteristics.

$$Y = \beta_0 + \sum_{i=1}^{4} \beta_i X_i + \sum_{i=1}^{6} \delta_{ij} X_i X_j + \varepsilon$$

This model focuses on interactions between variables without quadratic terms.

### 4. **Enhanced Linear Model with Targeted Interactions**

$$\begin{aligned}
Y = &\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_4 \\
&+ \delta_1 (X_1 \times X_3) + \delta_2 (X_1 \times X_4) \\
&+ \delta_3 (X_2 \times X_3) + \delta_4 (X_2 \times X_4) \\
&+ \delta_5 (X_3 \times X_4) + \delta_6 (X_1 \times X_2) + \varepsilon
\end{aligned}$$

Specific interaction terms:
- $X_1 \times X_3$ = `INT2K × green_view_ratio` (spatial integration × green view ratio)
- $X_1 \times X_4$ = `INT2K × sky_visibility` (spatial integration × sky visibility)
- $X_2 \times X_3$ = `CH2K × green_view_ratio` (choice × green view ratio)
- $X_2 \times X_4$ = `CH2K × sky_visibility` (choice × sky visibility)
- $X_3 \times X_4$ = `green_view_ratio × sky_visibility` (visual elements interaction)
- $X_1 \times X_2$ = `INT2K × CH2K` (spatial complexity interaction)

### **Variable Notation**

#### Original Variables:
- **INT2K**: Integration-2km, spatial integration values within 2km radius
- **CH2K**: Choice-2km, choice values within 2km radius  
- **green_view_ratio**: Green view ratio, proportion of green vegetation in street view images (0-1)
- **sky_visibility**: Sky visibility, proportion of sky in street view images (0-1)
- **safer_Trueskill_Scores**: Safety perception scores based on TrueSkill algorithm

#### Mathematical Symbols:
- $Y$: dependent variable (safety perception scores)
- $X_i$: i-th independent variable (standardized)
- $\beta_0$: intercept term (model constant)
- $\beta_i$: linear effect coefficients
- $\gamma_i$: quadratic effect coefficients (non-linear)
- $\delta_{ij}$: interaction effect coefficients
- $\varepsilon$: random error term
- $n$: sample size
- $\sigma^2$: error variance

### **Model Evaluation Metrics**

#### Goodness of Fit:
$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum(y_i - \hat{y_i})^2}{\sum(y_i - \bar{y})^2}$$

#### Adjusted R²:
$$R_{adj}^2 = 1 - \frac{(1-R^2)(n-1)}{n-p-1}$$

Where $p$ is the number of predictor variables.

#### Information Criteria:
- **AIC**: $AIC = 2p - 2\ln(L)$
- **BIC**: $BIC = p\ln(n) - 2\ln(L)$

Where $L$ is the maximum likelihood function value.

#### Prediction Error Metrics:
- **RMSE**: $RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y_i})^2}$
- **MAE**: $MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y_i}|$
- **MAPE**: $MAPE = \frac{1}{n}\sum_{i=1}^{n}\left|\frac{y_i - \hat{y_i}}{y_i}\right| \times 100\%$

### **Statistical Hypothesis Testing**

#### Overall Model Significance (F-test):
- $H_0$: $\beta_1 = \beta_2 = \beta_3 = \beta_4 = 0$
- $H_1$: At least one $\beta_i \neq 0$

$$F = \frac{MS_{regression}}{MS_{error}} = \frac{SS_{regression}/p}{SS_{error}/(n-p-1)}$$

#### Individual Coefficient Significance (t-test):
- $H_0$: $\beta_i = 0$
- $H_1$: $\beta_i \neq 0$

$$t = \frac{\hat{\beta_i}}{SE(\hat{\beta_i})}$$

---

Through these mathematical expressions, we can clearly understand how each model captures different types of relationships:
- **Linear Model**: Captures basic additive effects
- **Polynomial Model**: Additionally captures curvilinear relationships and interaction effects  
- **Interaction Model**: Focuses on interactions between variables
- **Enhanced Model**: Combines theoretically-driven specific interaction terms
